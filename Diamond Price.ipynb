```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import seaborn as sns
from IPython.display import HTML
```

```python
data = pd.read_csv("Resource/diamonds.csv")
data.head()
```

```python
data.info()
```

```python
diamond_data = data[['carat','cut','color','clarity','depth','table','price','x','y','z']]
diamond_data.describe()
```

```python
diamond_data = diamond_data.loc[(diamond_data.x!=0) & (diamond_data.y!=0) & (diamond_data.z!=0)]
diamond_data.shape
```

```python
diamond_data.isna().sum()
```

```python
# Drop the outliers 
diamond_data = diamond_data[(diamond_data.depth<75)&(diamond_data.depth>50)&(diamond_data.table<80)&(diamond_data.table>45)&(diamond_data.y<20)&(diamond_data.z<10)&(diamond_data.z>2)]
```

### Carat vs. Price

```python
x_values = diamond_data.carat
y_values = diamond_data.price

plt.scatter(x_values,y_values,facecolor='navy', alpha=0.6)
plt.rcParams["figure.figsize"] = (15,8)
plt.title('Diamond Carat vs. Price')
plt.xlabel('Diamond Carat')
plt.ylabel('Diamond Price')

# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
# regress_values = x_values * slope + intercept
# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# plt.plot(x_values,regress_values,"r-")
# plt.annotate(line_eq,(2500,4),fontsize=15,color="grey")

# print(f"The r value is: {rvalue}")
# print(f"The r-squared is: {rvalue**2}")
plt.savefig('image/Diamond Carat vs. Price.jpg')
plt.show()
```

```python
x_values = diamond_data.depth
y_values = diamond_data.price

plt.scatter(x_values,y_values,facecolor='navy', alpha=0.6)
plt.rcParams["figure.figsize"] = (15,8)
plt.title('Diamond Total Depth Percentage vs. Price')
plt.xlabel('Diamond Total Depth Percentage')
plt.ylabel('Diamond Price')

# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
# regress_values = x_values * slope + intercept
# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# plt.plot(x_values,regress_values,"r-")
# plt.annotate(line_eq,(2500,4),fontsize=15,color="grey")

# print(f"The r value is: {rvalue}")
# print(f"The r-squared is: {rvalue**2}")
plt.savefig('image/Diamond Total Depth Percentage vs. Price.jpg')
plt.show()
```

### Dimensions vs. Price
